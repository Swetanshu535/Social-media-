Question 1: What is the distribution of daily time spent on different platforms?

```python
df=pd.read_csv("C:\\Users\\Swetanshu\\Desktop\\Social_media .csv",usecols=['Platform','Daily Time Spent (min)'])
df=df.drop_duplicates(subset='Platform')
total_time=df['Daily Time Spent (min)'].sum()
plt.figure(figsize=(8,10))
plt.pie(df['Daily Time Spent (min)'],labels=df['Platform'],autopct='%1.1f%%',startangle=140)
plt.title("daily time spent and Platform")
plt.axis('equal')
plt.show
```

Question 2: How many platforms have verified accounts?

```python
df=pd.read_csv("C:\\Users\\Swetanshu\\Desktop\\Social_media .csv",usecols=['Platform','Verified Account'])
df=df.drop_duplicates(subset='Platform')
verified_counts = df['Verified Account'].value_counts()
plt.figure(figsize=(8,9))
plt.pie(verified_counts, labels=verified_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Verified account and Platform')
plt.axis('equal')
plt.show
```
